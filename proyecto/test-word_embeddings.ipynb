```python
import os

data_dir = "."
model_file = 'my_doc2vec.model'
train_file = 'verbalAutopsy_train.csv'
test_file = 'verbalAutopsy_test.csv'
```

```python
# Inicializamos y entrenamos un modelo Doc2Vec
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import pandas as pd

#documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(common_texts)]
train_df = pd.read_csv(train_file)
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(train_df['open_response'])]

model = Doc2Vec(documents, vector_size=7, window=2, min_count=1, workers=4)

# Guardamos el modelo
model.save(model_file)
```

```python
from gensim.utils import simple_preprocess
from gensim.models.doc2vec import Doc2Vec
import gensim.downloader
import smart_open

# Cargamos el modelo
model = Doc2Vec.load(model_file)
# model = gensim.downloader.load('glove-twitter-25')

def read_corpus(fname, tokens_only=False):
    with smart_open.open(fname, encoding="iso-8859-1") as f:
        for i, line in enumerate(f):
            # Preprocesado del documento
            tokens = simple_preprocess(line)

            # Vectorización
            if tokens_only:
                yield tokens
            else:
                # Vectorizamos con Doc2Vec
                #yield gensim.models.doc2vec.TaggedDocument(tokens, [i])
                yield model.infer_vector(tokens)

train_corpus = list(read_corpus(train_file))
test_corpus = list(read_corpus(test_file, tokens_only=True))
```
